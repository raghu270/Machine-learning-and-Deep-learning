```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris.feature_names
```

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
data.head()
```

```python
data['target']=iris.target
```

```python
iris.target_names
```

```python
data
```

```python
data["target"]==1
```

```python
data[data["target"]==1]
```

```python
data["flower_name"]=data.target.apply(lambda x:iris.target_names[x])
```

```python
data
```

```python
data1=data[data["target"]==0]
data2=data[data["target"]==1]
data3=data[data["target"]==2]
```

```python
data1
```

```python
plt.scatter(data1["sepal length (cm)"],data1["sepal width (cm)"])
plt.scatter(data2["sepal length (cm)"],data2["sepal width (cm)"])
plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

```python
plt.scatter(data1["petal length (cm)"],data1["petal width (cm)"])
plt.scatter(data2["petal length (cm)"],data2["petal width (cm)"])
plt.xlabel('petal length')
plt.ylabel('petal width')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop(["target","flower_name"],axis="columns")
```

```python
x
```

```python
y=data.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.svm import SVC
model=SVC(kernel="linear")
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3.1,4.1,3.2,4.2]])
```
