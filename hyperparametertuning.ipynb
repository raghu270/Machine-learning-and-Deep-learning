```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.data
```

```python
iris.feature_names
```

```python
data=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
data
```

```python
data["type"]=iris.target
```

```python
data
```

```python
iris.target_names[0]
```

```python
data["type"]=data["type"].apply(lambda x:iris.target_names[x])
```

```python
data["type"]
```

```python
data
```

```python

x_train,x_test,y_train,y_test=train_test_split(data.drop("type",axis="columns"),data["type"],test_size=0.2)
```

```python
x_train
```

```python
data["type"]
```

```python
data.drop("type",axis="columns")
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
clf=GridSearchCV(SVC(gamma='auto'),{'C':[1,10,20],'kernel':['rbf','linear']},cv=5,return_train_score=False)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
data2=pd.DataFrame(clf.cv_results_)
```
