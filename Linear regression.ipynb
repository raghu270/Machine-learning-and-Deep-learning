```python
import numpy as np
import matplotlib.pyplot as plt
def coef(x,y):
    n=np.size(x)
    m_x=np.mean(x)
    m_y=np.mean(y)
    ss_xy=np.sum(x*y)-n*m_x*m_y
    ss_xx=np.sum(x*x)-n*m_x*m_x
    b_1=ss_xy/ss_xx
    b_0=y[0]-b_1*x[0]
    return (b_0,b_1)
def plot_reg(x,y,b):
    plt.scatter(x,y,marker='o',s=30)
    y_pred=b[0]+b[1]*x
    plt.plot(x,y_pred)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
x=np.array([2,4,6,8])
y=np.array([3,7,5,10])
b=coef(x,y)
print("The coeffiecients are b_0={},b_1={}".format(b[0],b[1]))
print(b[0],b[1])
plot_reg(x,y,b)
```
